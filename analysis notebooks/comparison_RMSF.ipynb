```python
#creating an average structure

average = align.AverageStructure(u, u, select='resid 0:12 and not name H*',
                                 ref_frame=0).run()
ref = average.results.universe

average_2 = align.AverageStructure(u_2, u_2, select='resid 0:12 and not name H*',
                                 ref_frame=0).run()
ref_2 = average_2.results.universe

average_3 = align.AverageStructure(u_3, u_3, select='resid 0:12 and not name H*',
                                 ref_frame=0).run()
ref_3 = average_3.results.universe

average_4 = align.AverageStructure(u_4, u_4, select='index 0:92',
                                 ref_frame=0).run()
ref_4 = average_4.results.universe

average_5 = align.AverageStructure(u_5, u_5, select='index 0:92',
                                 ref_frame=0).run()
ref_5 = average_5.results.universe

average_6 = align.AverageStructure(u_6, u_6, select='index 0:92',
                                 ref_frame=0).run()
ref_6 = average_6.results.universe
```

```python
aligner = align.AlignTraj(u, ref,
                          select='resid 0:12 and not name H*',
                           filename='aligned_traj.dcd',
                           in_memory=False).run()
u = mda.Universe("/pascal/data/project/100ns_prod_runs/prod_100ns_amber_4000_1/amber_copy.parm7", 'aligned_traj.dcd')

aligner_2 = align.AlignTraj(u_2, ref_2,#amber trajectories
                          select='resid 0:12 and not name H*',
                           filename='aligned_traj_2.dcd',
                           in_memory=False).run()
u_2 = mda.Universe("/pascal/data/project/100ns_prod_runs/prod_100ns_amber_4000_2/amber_copy.parm7", 'aligned_traj_2.dcd')

aligner_3 = align.AlignTraj(u_3, ref_3,
                          select='resid 0:12 and not name H*',
                           filename='aligned_traj_3.dcd',
                           in_memory=False).run()
u_3 = mda.Universe("/pascal/data/project/100ns_prod_runs/prod_100ns_amber_4000_3/amber_copy.parm7", 'aligned_traj_3.dcd')



aligner_4 = align.AlignTraj(u_4, ref_4,
                          select='index 0:92',
                           filename='aligned_traj_4.dcd',
                           in_memory=False).run()
u_4 = mda.Universe("/pascal/data/project/protein_ligand/prod_100ns_esp_4000/amber_copy.parm7", 'aligned_traj_4.dcd')

aligner_5 = align.AlignTraj(u_5, ref_5,
                          select='index 0:92',
                           filename='aligned_traj_5.dcd',
                           in_memory=False).run()
u_5 = mda.Universe("/pascal/data/project/protein_ligand/prod_100ns_esp_4000_2/amber_copy.parm7", 'aligned_traj_5.dcd')

aligner_6 = align.AlignTraj(u_6, ref_6,
                          select='index 0:92',
                           filename='aligned_traj_6.dcd',
                           in_memory=False).run()
u_6 = mda.Universe("/pascal/data/project/protein_ligand/prod_100ns_esp_4000_3/amber_copy.parm7", 'aligned_traj_6.dcd')
```

```python
heavy_atoms = u.select_atoms('resid 0:12 and not name H*')
heavy_atoms_2 = u_2.select_atoms('resid 0:12 and not name H*')
heavy_atoms_3 = u_3.select_atoms('resid 0:12 and not name H*')
heavy_atoms_4= u_4.select_atoms('index 0:92')
heavy_atoms_5= u_5.select_atoms('index 0:92')
heavy_atoms_6= u_6.select_atoms('index 0:92')

R = rms.RMSF(heavy_atoms).run()
R_2 = rms.RMSF(heavy_atoms_2).run()
R_3 = rms.RMSF(heavy_atoms_3).run()
R_4 = rms.RMSF(heavy_atoms_4).run()
R_5 = rms.RMSF(heavy_atoms_5).run()
R_6 = rms.RMSF(heavy_atoms_6).run()
```
